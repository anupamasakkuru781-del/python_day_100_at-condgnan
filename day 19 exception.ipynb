```python
try:
    a = 10
    b = 0
    c = a / b
    print(c)
except ZeroDivisionError:
       print("zero Division not possible")
except:
    print("Something wrong in logic")
```

```python
# excute the coresponding  except block code
try:
    a = 10
    c = a + 'hi'
    print(c)
except ZeroDivisionError:
       print("zero Division not possible")
except:
    print("Something wrong in logic")
```

```python
# excute the corescponding except block code
try:
    a = 10
    c = a + 'hi'
    print(c)
except ZeroDivisionError:
       print("zero Division not possible")
except TypeError:
    print("concatenation not possiable between int and str")
```

```python
# excute the corescponding except block code
try:
    a = 10
    c = a.append(10)
    print(c)
except ZeroDivisionError:
       print("zero Division not possible")
except TypeError:
    print("concatenation not possiable between int and str")
except Exception as e:
    print(e)
```

```python
# multiple Eception in single except statement
try:
    a = [10,20,30]
    c = a[10]
    print(c)
except ZeroDivisionError:
       print("zero Division not possible")
except TypeError:
    print("concatenation not possiable between int and str")
except Exception as e:
    print(e)
```
