```python
class Contact:
    def __init__(self, name, number):
        self.name = name
        self.number = number

    def update(self, number):
        self.number = number
        print("Contact updated successfully")


class ContactBook:
    def __init__(self):
        self.contacts = []

    def add_contact(self, contact):
        self.contacts.append(contact)
        print("Contact added successfully")

    def search_contact(self, name):
        for contact in self.contacts:
            if contact.name == name:
                print("Contact found:", contact.name, contact.number)
                return contact
        print("Contact not found")
        return None

    def delete_contact(self, name):
        contact = self.search_contact(name)
        if contact:
            self.contacts.remove(contact)
            print("Contact deleted successfully")
        else:
            print("Contact not found")


# -------- Main Program --------
contact_book = ContactBook()

contact1 = Contact("ravi", 9846442450)
contact2 = Contact("shiva", 8011632531)


contact_book.add_contact(contact1)
contact_book.add_contact(contact2)


print("\nadd contact")
print("update contact")
print("search contact")
print("delete contact")

contact_book.search_contact("ravi")
contact1.update(9999999999)
contact_book.delete_contact("shiva")
```
