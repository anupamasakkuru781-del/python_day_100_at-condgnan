```python
# book managing section
books = {}
def add_book():
    book_id = int(input("enter book ID:"))
    name = (input("enter name of the book:"))
    author = (input("enter the author name:"))
    quantity = int(input("enter quantity:"))
    books[book_id] = [name, author, quantity]
    print("Book added successfully")
def delete_book():
    book_id = int(input("enter the ID:"))
    if book_id in  books:
        del books[book_id]
        print("Book deleted")
    else:
        print("Book not found")
def total_books():
    print("Total number of books:", len(books))
def all_books():
    if not books:
        print("books  not in available")
        return
    print("enter all books")
    for book_id, details in books.items():
        print(f"ID: {book_id}, Name: {details[0]}, Author: {details[1]}, Quantity: {details[2]}")
def barrow_book():
    book_id = int(input("enter book ID number:"))
    if book_id in books:
        if books[book_id][2] > 0:
            books[book_id][2] -= 1
            print("book borrowed")
    else:
        print("book not found")
while True:
    print("1. Add book")
    print("2. Delete book")
    print("3. Total book")
    print("4. all books")
    print("5. borrow book")
    choice = int(input("enter your choice:"))
    if choice == 1:
        add_book()
    elif choice == 2:
        delete_book()
    elif choice == 3:
        total_books()
    elif choice == 4:
         all_books()
    elif choice == 5:
        barrow_book()
        print("program exits")
        break
    else:
        print("book details not exsisting")
```
