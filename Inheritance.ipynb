```python
# Parent class
class User:
    def __init__(self, username, email, password, age):
        self.username = username
        self.email = email
        self.password = password
        self.age = age

    def details(self):
        return f"My name is {self.username}, email id is {self.email} and I am {self.age} years old"


# Child class
class Admin(User):
    def __init__(self, username, email, password, age, total_users):
        super().__init__(username, email, password, age)
        self.total_users = total_users

    def show_admin_details(self):
        return f"I am working as admin and total users are {self.total_users}"


# Object creation
u1 = User(username="babu", email="babu@gmail.com", password="babu123", age=30)
a1 = Admin(username="srinu", email="srinu@gmail.com", password="srinu@123", age=35, total_users=100)

# Output
print("User details:", u1.details())
print("Admin username:", a1.username)
print("Admin details:", a1.details())
print("Admin extra info:", a1.show_admin_details())
```
