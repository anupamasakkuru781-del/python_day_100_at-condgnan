```python
# method overriding
# person
class Person():
     def __init__(self, name, age):
         self.name = name
         self.age = age

     def show(self):
        return f"Myself {self.name} and i am a {self.age} years old"
         

# student
class Student(person):
    def __init__(self, name, age, std):
        print(super().__init__(name, age))
        person.__init__(self, name, age)
        self.std = std

    def show(self):
        return f"Myself {self.name} and i am studing {self.std} class"
p = Person('Babu',50)
s = Student('Srinu',15,10)
print(p.show())
print(s.show())
```

```python
# method overloading
class Math:
    def add(self, a, b):
        return a + b
    def add(self, a, b, c):
        return a + b + c
    def add(self, a, b, c, d):
        return a + b + c + d

m = Math()
print(m.add(1,2,3,4))
print(m.add(1,2,3))
print(m.add(1,2))
```

```python
# method overloading achieve by using defualt arguments
class Math:
    def add(self, a, b, c=0, d=0):
        return a + b + c + d

m = Math()
print(m.add(1,2,3,4))
print(m.add(1,2,3))
print(m.add(1,2))
```
