```python
# create student class:
class Student:
    school = "xyz school"

    def __init__(self, name, age, cls):
        self.name = name
        self.age = age
        self.cls = cls

    def details(self):
        return f"My name is {self.name}, and {self.age} years old and belong to class {self.cls}"


s1 = Student("Ravi", 14, 9)
s2 = Student("Babu", 15, 10)

print(s1.name)
print(s2.name)
print(s1.details())
print(s2.details())
print(s1.school)
print(s2.school)
```
