```python
# variable length of positional arguments
def add(a, *args):
        print(a)
        print(args)
add(1,2,3,4,5,6,7,8,9)
```

```python
# variable length of positional arguments
def add(a, *args):
        print(a)
        print(args)
        s=0
        for num in args:
            s += num
            return s
add(1,2,3,4,5,6)
```
