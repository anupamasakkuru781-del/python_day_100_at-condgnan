```python
# set allows unique elements
# set unordered
s1 = {1,2,3,4,'hi',1,2,3,'a'}
print(s1)
```

```python
# set not allow list as element
s1 = {1,2,3,4,'hi',1,2,3,'a',[1,2,3,4]}
print(s1)
```

```python
# set not allow dictionary as element
s1 = {1,2,3,4,'hi',1,2,3,'a',{1:2}}
print(s1)
```

```python
## set functions
# adding set elements to set
#add(ele): it add the element to set if element is not in set otherwise nothing
# update(iterator): it add the all element in iterator
```

```python
s1 = {1,2,3,4}
s1.add(10)
s1.add(1)
print(s1)
s1.update([5,3,5])
print(s1)
```

```python
s1 = {1,2,3,4,10}
s1.remove(10)
print(s1)
```

```python
## remove elements from set
# remove(): It removes the elements from set if its present in set otherwise keyerror
#discard(): It removes the elements from set if its present in set otherwise nothing
#pop(): It removes and returns the radom elements from set
s1 = {1,2,3,4,10}
s1.discard(2)
print(s1)
```

```python
s1 = {1,2,3,4,10}
s1.add(15)
print(s1)
s1.discard(2)
print(s1)
```

```python
s1 = {1,2,3,4,10}
print(s1.pop())
print(s1)
```
