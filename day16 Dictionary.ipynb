```python
# deleting item from dictionary
# pop(key, default) - It return value and removes the specified key item from dictionary
# popitem() - It return and removes the last item from dictionary
details = {'name': 'babu', 'batch': 47, 'course': 'python', 'age': 50}
popped_value = details['age']
print(popped_value)
print(details)
popped_value = details.pop('age')
print(details)
```
